```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt

from numpy.random import randint
from numpy import mean
from scipy import stats
```

Create a line chart plotting the sample mean as more and more trials are conducted, up to 1,000 trials.

# Normal Distribution 

```python
trials = []

for rand in range(1, 1000):
    mean_results = np.random.normal(1,10,20).mean()
    trials.append(mean_results)

df = pd.DataFrame({'trials': trials})
df.plot()
```

Randomly sample the mean 1,000 times and plot the resulting frequency of sample means as a histogram.

```python
norm_ = [mean(randint(1, 10, 20)) for rand in range(1, 1000)]
```

```python
plt.hist(norm_)
print(stats.describe(norm_))
```

# Binomial

```python
trials = []

for rand in range(1,10000):  
    bin_ = np.random.binomial(1, .5, 100).mean()
    trials.append(bin_)

df_bino = pd.DataFrame(trials)
df_bino.plot()
```

```python
np.mean(trials)
```

```python
bino_ = [mean(randint(1, 10, 100)) for rand in range(1, 1000)]
```

```python
plt.hist(bino_)
```
