Hepatitis B, HIV/AIDS, GDP has the most significant outliers. 

We will verify this by running the Tukey method. By brining the method into the first 25 and max 75 percentiles. We will also transform the the outliers using log transformation and see if it has an effect on the outliers.

```python
#Hepatitis B - Identifying Outliers
q75, q25 = np.percentile(lifeexpectancy_df['Hepatitis B'], [75 ,25])
iqr = q75 - q25

for threshold in np.arange(1,5,0.5):
    min_val = q25 - (iqr*threshold)
    max_val = q75 + (iqr*threshold)
    print("The score threshold is: {}".format(threshold))
    print("Number of outliers is: {}".format(
        len((np.where((lifeexpectancy_df['Hepatitis B'] > max_val) 
                      | (lifeexpectancy_df['Hepatitis B'] < min_val))[0]))
    ))
```

```python
#Hepatitis B Treating outliers via Winsorize
plt.figure(figsize = (15,5))
winsorized_views_HepB = winsorize(lifeexpectancy_df['Hepatitis B'], (0, 0.10))

plt.subplot(1,2,1)
plt.boxplot(lifeexpectancy_df['Hepatitis B'])
plt.title('Box plot of Hepatitis B')

plt.subplot(1,2,2)
#plt.boxplot(np.log(lifeexpectancy_df['Hepatitis B']))
plt.boxplot(winsorized_views_HepB)
plt.title('Box plot of Hepatitis B (whis=1.5)')

plt.show()
```

```python
#HIV/AIDS - Identifying outliers
q75, q25 = np.percentile(lifeexpectancy_df['HIV/AIDS'], [75 ,25])
iqr = q75 - q25

for threshold in np.arange(1,5,0.5):
    min_val = q25 - (iqr*threshold)
    max_val = q75 + (iqr*threshold)
    print("The score threshold is: {}".format(threshold))
    print("Number of outliers is: {}".format(
        len((np.where((lifeexpectancy_df['HIV/AIDS'] > max_val) 
                      | (lifeexpectancy_df[' HIV/AIDS'] < min_val))[0]))
    ))
```

```python
#HIV/Aids Treating outliers via Winsorize
plt.figure(figsize = (15,5))
winsorized_views_HIVAIDS = winsorize(lifeexpectancy_df['HIV/AIDS'], (0, 0.10))

plt.subplot(1,2,1)
plt.boxplot(lifeexpectancy_df['HIV/AIDS'])
plt.title('Box plot of HIV/AIDS')

plt.subplot(1,2,2)
plt.boxplot(winsorized_views_HIVAIDS)
plt.title('Box plot of HIV/AIDS (whis=1.5)')

plt.show()
```

```python
#GDP - Identifying Outliers
q75, q25 = np.percentile(lifeexpectancy_df['GDP'], [75 ,25])
iqr = q75 - q25

for threshold in np.arange(1,5,0.5):
    min_val = q25 - (iqr*threshold)
    max_val = q75 + (iqr*threshold)
    print("The score threshold is: {}".format(threshold))
    print("Number of outliers is: {}".format(
        len((np.where((lifeexpectancy_df['GDP'] > max_val) 
                      | (lifeexpectancy_df['GDP'] < min_val))[0]))
    ))
```

```python
#GDP Treating outliers via Winsorize
plt.figure(figsize = (15,5))
winsorized_views_GDP = winsorize(lifeexpectancy_df['GDP'], (0, 0.10))

plt.subplot(1,2,1)
plt.boxplot(lifeexpectancy_df['GDP'])
plt.title('Box plot of GDP')

plt.subplot(1,2,2)
plt.boxplot(winsorized_views_GDP)
plt.title('Box plot of GDP (whis=1.5)')

plt.show()
```
