#### Filling Average Math / Reading Scores with the mean

```python
#Math 4 fill missing with mean
useducation_df['AVG_MATH_4_SCORE'].fillna(
    useducation_df['AVG_MATH_4_SCORE'].mean(), inplace=True)

#Math 8 fill missing with mean
useducation_df['AVG_MATH_8_SCORE'].fillna(
    useducation_df['AVG_MATH_8_SCORE'].mean(), inplace=True)

#Reading 4 fill missing with mean
useducation_df['AVG_READING_4_SCORE'].fillna(
    useducation_df['AVG_READING_4_SCORE'].mean(), inplace=True)

#Reading 8 fill missing with mean
useducation_df['AVG_READING_8_SCORE'].fillna(
    useducation_df['AVG_READING_8_SCORE'].mean(), inplace=True)
```
