```python
x = useducation_df['GRADES_8_G'] - useducation_df['GRADES_4_G']
x.sum()
```

```python
cols = ['ENROLL', 'TOTAL_REVENUE', 'FEDERAL_REVENUE','STATE_REVENUE', 
        'LOCAL_REVENUE', 'TOTAL_EXPENDITURE', 'INSTRUCTION_EXPENDITURE', 
        'SUPPORT_SERVICES_EXPENDITURE','OTHER_EXPENDITURE', 
        'CAPITAL_OUTLAY_EXPENDITURE', 'GRADES_PK_G','GRADES_KG_G', 
        'GRADES_4_G', 'GRADES_8_G', 'GRADES_12_G', 'GRADES_1_8_G',
        'GRADES_9_12_G', 'GRADES_ALL_G']

for state in cols:
    useducation_df.loc[useducation_df['STATE'] == state, cols ] = useducation_df.loc[useducation_df['STATE'] == state, cols ].mean()

useducation_df.dropna(inplace=True)
```

#### Compute the average score for each row for all students weighted properly by grade. Notice that the number of students in the 4th grade isn't the same as the number of students in the 8th grade. So, you should appropriately weigh the scores.


```python
useducation_df['total_score'] = (useducation_df['GRADES_4_G']*((useducation_df['AVG_MATH_4_SCORE'] + useducation_df['AVG_READING_4_SCORE'])*0.5) + useducation_df['GRADES_8_G']
                                 * ((useducation_df['AVG_MATH_8_SCORE'] + useducation_df['AVG_READING_8_SCORE'])*0.5))/(useducation_df['GRADES_4_G'] + useducation_df['GRADES_8_G'])
```

#### What are the correlations between this newly created score variable and the expenditure types? Which 1 of the expenditure types is more correlated than the others?


```python
useducation_df[['total_score', 'TOTAL_EXPENDITURE', 'INSTRUCTION_EXPENDITURE', 'SUPPORT_SERVICES_EXPENDITURE',
               'OTHER_EXPENDITURE','CAPITAL_OUTLAY_EXPENDITURE']].corr()
```

Instruction expenditure is more correlated than the others

#### Now, apply PCA to the 4 expenditure types. How much of the total variance is explained by the 1st component?

```python
X = useducation_df[["INSTRUCTION_EXPENDITURE", "SUPPORT_SERVICES_EXPENDITURE",
                  "OTHER_EXPENDITURE", "CAPITAL_OUTLAY_EXPENDITURE"]]

X = StandardScaler().fit_transform(X)

sklearn_pca = PCA(n_components=1)
useducation_df["pca"] = sklearn_pca.fit_transform(X)

pca_ = sklearn_pca.explained_variance_ratio_
print(pca_)
```

more than 94% of the total cariance is expland by each componment from Sklearn PCA
