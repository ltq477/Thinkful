```python
import pandas as pd
import seaborn as sns
from scipy import stats
sns.set()

house_prices_df = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/houseprices.csv')
```

```python
house_prices_df.info()
```

Illustrate the distribution of saleprice.

```python
saleprice = sns.distplot(house_prices_df['saleprice'], kde=False)
```

Illustrate the distribution of the lotarea.


```python
lotarea = sns.distplot(house_prices_df['lotarea'], kde=False)
```

Use a boxplot to illustrate how overallcond affects the sale price.


```python
over_sales = sns.boxplot(x='overallcond', 
                         y='saleprice',
                         showfliers=False,
                         data=house_prices_df);

over_sales.set(xlabel = 'Overall Condition Rating',
              ylabel = 'Sales Price');
```

On the same boxplot, highlight the difference between having central air conditioning or not.


```python
over_sales = sns.boxplot(x='overallcond', 
                         y='saleprice', hue='centralair', 
                         showfliers=False,
                         data=house_prices_df);
over_sales.set(xlabel = 'Overall Condition Rating',
              ylabel = 'Sales Price');
```

# Is there a correlation between lotarea and the saleprice?

There is no correlation as illustrated below. We see a flat line with most of the data in the beginning location. The increase in lot area cannot directly translate to the increase in sales price.

```python
sns.scatterplot(y='lotarea', x='saleprice', data=house_prices_df);
```

```python
stats.pearsonr(house_prices_df['lotarea'], house_prices_df['saleprice'])
```

```python
sns.boxplot(x = 'lotarea', y='saleprice', fliersize=False, whis=4, data=house_prices_df);
```
