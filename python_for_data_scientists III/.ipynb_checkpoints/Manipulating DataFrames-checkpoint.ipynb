This is a dataset consisting of statistics about Kickstarter projects. Download the dataset here. Create a DataFrame from the table named ksprojects. Write code that allows you to answer the following questions. Create a separate cell for each question.

```python
import pandas as pd
import numpy as np 

kickstarter_data = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/ksprojects.csv')
```

Find all successful documentary projects and sort them by the amount pledged. Print the top 10 highest pledges.

```python
#kickstarter_data.sort_values(by=(kickstarter_data['category' == 'Documentary'] & kickstarter_data['state' == 'successful']))
state = kickstarter_data.iloc[(kickstarter_data['state'] == 'successful').values] 
cat = kickstarter_data.iloc[(kickstarter_data['category'] == 'Documentary').values]

state_cat = pd.DataFrame(pd.merge(state, cat))
pledged_top = (state_cat.iloc[(state_cat['pledged'] != 0).values])
kickstarter_data = pledged_top.sort_values('pledged', ascending=False)

print('The top 10 highest successful documentary projects are the ff:')
kickstarter_data.head(10)
```

```python
sd = kickstarter_data[
    (kickstarter_data['state'] == 'successful') & 
    (kickstarter_data['category'] == 'Documentary')
]

sd.sort_values('pledged', ascending=False).head(10)
```

Create a new column named average_per_backer and set its value to the total amount pledged / number of backers. 

```python
#created new column called 'average_per_backer'
kickstarter_data = kickstarter_data.assign(average_per_backer = 
                                             (kickstarter_data.pledged) / (kickstarter_data.backers)
                                            )
print('The new column list which includes the new column are the ff:')
kickstarter_data.columns
```

```python
print('showing the head of the DataFrame with the new column')
kickstarter_data.head()
```

What happened to the rows with 0 backers? How can this be dealt with?

```python
print('The rows with 0 backers initially remained but were removed accordingly')
```

Drop all rows with 0 backers then repeat the previous exercise.

```python
#drop rows with zero backers
kickstarter_data = kickstarter_data[kickstarter_data.backers != 0]

print('The new DataFrame without rows with 0 backers')
kickstarter_data.head()
```

```python
#check if any 0 backers in the DataFrame
print('Checking the DataFrame is there are any 0 backers')
kickstarter_data[kickstarter_data.backers == 0]
```

```python
# Create new column referencing the last dataframe with 0 backers
kickstarter_df = kickstarter_data.assign(average_per_backer = (kickstarter_data.pledged) / (kickstarter_data.backers))

#list name of all columns including new column 'average_per_backer' 
kickstarter_df.columns
```

Create a crosstab to get a count of records for each combination of state and category.

```python
pd.crosstab(kickstarter_df['state'], kickstarter_df['category'])
```

# census

```python
population = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/state-populations.csv')
region = pd.read_csv('https://tf-assets-prod.s3.amazonaws.com/tf-curric/data-science/census-divisions.csv')
```

Merge these two DataFrames into a single DataFrame called census

```python
census = pd.merge(population, region)
census.head(5)
```

Re-shape census so that one column contains all population measures, and another the year attributes.

```python
census.columns
```

```python
census = pd.melt(frame=census, id_vars=('state','region','division'), var_name='year')
census
```

Group the data by year and summarize it.

```python
census.groupby(['year']).mean()
```

Group the data by region, division and year and summarize it.

```python
census.groupby(['region','division','year']).describe()
```
